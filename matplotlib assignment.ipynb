# 2.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

scatter plot provides a scattered relation between two data represented in two axis

```python
plt.scatter(x,y)
plt.xlabel("X_AXIS")
plt.ylabel("Y_AXIS")
plt.title("SCATTER_PLOT")
```

# 3.

subplot is used to make multiple plots at a time.

```python
x = np.array([0, 1, 2, 3, 4, 5])  
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
```

# 4.

bar plot is used to represents data in the form vartical columns

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
```

# 5.

Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.subplot(2,2,1)
plt.boxplot(box1)
plt.xlabel("BOX1")
plt.subplot(2,2,2)
plt.boxplot(box2)
plt.xlabel("BOX2")
```
